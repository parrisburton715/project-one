```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json
```

```python
# Create a file path
data_set = Path("Maternal Death Rate.csv")
# Read in the data. 
maternal_death_df = pd.read_csv(data_set)
maternal_death_df.head()
```

```python
# Plotting the bar chart(Etnicity vs Maternal Deaths)
plt.figure(figsize=(12, 6))
plt.bar(maternal_death_df["Ethnicity"], maternal_death_df["Maternal Deaths"], color='skyblue')
plt.title('Maternal Death Par Race (2019 - 2023)')
plt.xlabel('Ethnicity')
plt.ylabel('Maternal Deaths')
plt.xticks(rotation=45, ha='right')  # Rotate state names for better readability
plt.tight_layout()
plt.show()
```

```python
# Plotting the bar chart(Ethnicity Vs Live Births)
plt.figure(figsize=(12, 6))
plt.bar(maternal_death_df["Ethnicity"], maternal_death_df["Live Births"], color='skyblue')
plt.title('')
plt.xlabel('Ethnicity')
plt.ylabel('Live Births')
plt.xticks(rotation=45, ha='right')  # Rotate state names for better readability
plt.tight_layout()
plt.show()
```

```python
# Plotting the bar chart(Ethnicity vs Maternal Death Rate)
plt.figure(figsize=(12, 6))
plt.bar(maternal_death_df["Ethnicity"], maternal_death_df["Maternal Death Rate(10000)"], color='red')
plt.title('Maternal Death Per Race (2019 - 2023)')
plt.xlabel('Ethnicity')
plt.ylabel('Maternal Death Rate (*10000)')
plt.xticks(rotation=45, ha='right')  # Rotate state names for better readability
plt.tight_layout()
plt.show()
```
